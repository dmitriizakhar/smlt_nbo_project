```python
 import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

df_campaigns = pd.read_feather("for_students/campaigns.feather")

df_campaigns['delivery_date'] = pd.to_datetime(df_campaigns['delivery_date'], errors='coerce')

print("Информация о df_campaigns:")
print(df_campaigns.info())
print("\nОписание (включая категориальные признаки):")
print(df_campaigns.describe(include='all'))

```

```python
# -----------------------------
# 1. Распределение по A/B-группам
# -----------------------------
plt.figure(figsize=(8, 5))
sns.countplot(data=df_campaigns, x='ab_group', palette="pastel")
plt.title("Распределение A/B-групп")
plt.xlabel("A/B группа")
plt.ylabel("Количество кампаний")
plt.show()

# -----------------------------
# 2. Распределение по segment_name
# -----------------------------
plt.figure(figsize=(10, 6))
order_seg = df_campaigns['segment_name'].value_counts().index
sns.countplot(data=df_campaigns, x='segment_name', order=order_seg, palette="Blues_d")
plt.title("Распределение по segment_name")
plt.xlabel("Segment Name")
plt.ylabel("Количество кампаний")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 3. Распределение по tag_category
# -----------------------------
plt.figure(figsize=(10, 6))
order_cat = df_campaigns['tag_category'].value_counts().index
sns.countplot(data=df_campaigns, x='tag_category', order=order_cat, palette="Set2")
plt.title("Распределение по tag_category")
plt.xlabel("Tag Category")
plt.ylabel("Количество кампаний")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 4. Распределение по tag_product
# -----------------------------
plt.figure(figsize=(10, 6))
order_prod = df_campaigns['tag_product'].value_counts().index
sns.countplot(data=df_campaigns, x='tag_product', order=order_prod, palette="Set3")
plt.title("Распределение по tag_product")
plt.xlabel("Tag Product")
plt.ylabel("Количество кампаний")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 5. Распределение по target_event
# -----------------------------
plt.figure(figsize=(10, 6))
order_event = df_campaigns['target_event'].value_counts().index
sns.countplot(data=df_campaigns, x='target_event', order=order_event, palette="Set1")
plt.title("Распределение по target_event")
plt.xlabel("Target Event")
plt.ylabel("Количество кампаний")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 6. Дополнительные распределения:
# tag_campaign_type, tag_discount_flag, tag_chain_flag, tag_model_flag, tag_realestate_flag
# -----------------------------
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

sns.countplot(data=df_campaigns, x='tag_campaign_type',
              order=df_campaigns['tag_campaign_type'].value_counts().index,
              palette="coolwarm", ax=axes[0,0])
axes[0,0].set_title("Распределение по tag_campaign_type")
axes[0,0].set_xlabel("Tag Campaign Type")
axes[0,0].set_ylabel("Количество кампаний")
axes[0,0].tick_params(axis='x', rotation=45)

sns.countplot(data=df_campaigns, x='tag_discount_flag',
              order=df_campaigns['tag_discount_flag'].value_counts().index,
              palette="pastel", ax=axes[0,1])
axes[0,1].set_title("Распределение по tag_discount_flag")
axes[0,1].set_xlabel("Tag Discount Flag")
axes[0,1].set_ylabel("Количество кампаний")
axes[0,1].tick_params(axis='x', rotation=45)

sns.countplot(data=df_campaigns, x='tag_chain_flag',
              order=df_campaigns['tag_chain_flag'].value_counts().index,
              palette="muted", ax=axes[1,0])
axes[1,0].set_title("Распределение по tag_chain_flag")
axes[1,0].set_xlabel("Tag Chain Flag")
axes[1,0].set_ylabel("Количество кампаний")
axes[1,0].tick_params(axis='x', rotation=45)

sns.countplot(data=df_campaigns, x='tag_model_flag',
              order=df_campaigns['tag_model_flag'].value_counts().index,
              palette="bright", ax=axes[1,1])
axes[1,1].set_title("Распределение по tag_model_flag")
axes[1,1].set_xlabel("Tag Model Flag")
axes[1,1].set_ylabel("Количество кампаний")
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# -----------------------------
# 7. Анализ по времени: распределение по delivery_date
# -----------------------------
plt.figure(figsize=(12, 6))
sns.histplot(data=df_campaigns, x='delivery_date', bins=30, kde=True)
plt.title("Распределение дат доставки кампаний")
plt.xlabel("Delivery Date")
plt.ylabel("Количество кампаний")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 8. Анализ распределения в разрезе A/B-групп
# Например, распределение target_event по A/B-группам:
plt.figure(figsize=(12, 6))
sns.countplot(data=df_campaigns, x='target_event', hue='ab_group',
              order=order_event, palette="Set2")
plt.title("Распределение target_event по A/B-группам")
plt.xlabel("Target Event")
plt.ylabel("Количество кампаний")
plt.xticks(rotation=45)
plt.legend(title="A/B группа")
plt.show()
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

df_targets = pd.read_feather("for_students/targets.feather")
df_targets['report_date'] = pd.to_datetime(df_targets['report_date'], errors='coerce')

plt.figure()
sns.countplot(data=df_targets, y='event',
              order=df_targets['event'].value_counts().index,
              palette="Set2")
plt.title("Распределение типов событий")
plt.xlabel("Количество")
plt.ylabel("Тип события")
plt.show()

user_counts = df_targets.groupby('event')['smt_decision_makerid'].nunique().reset_index()
user_counts = user_counts.rename(columns={'smt_decision_makerid': 'unique_users'})

plt.figure()
sns.barplot(data=user_counts, y='event', x='unique_users', palette="Set1",
            order=user_counts.sort_values('unique_users', ascending=False)['event'])
plt.title("Уникальные пользователи по типам событий")
plt.xlabel("Количество уникальных пользователей")
plt.ylabel("Тип события")
plt.show()
```

```python

# -----------------------------
# 3. Временной тренд: количество событий по report_date
# -----------------------------
events_over_time = df_targets.groupby(df_targets['report_date'].dt.date).size().reset_index(name='count')

plt.figure()
sns.lineplot(x='report_date', y='count', data=events_over_time, marker="o")
plt.title("Количество событий по датам")
plt.xlabel("Дата")
plt.ylabel("Количество событий")
plt.xticks(rotation=45)
plt.show()


# -----------------------------
# 5. Распределение по tag_event_project_type
# -----------------------------
plt.figure()
sns.countplot(data=df_targets, x='tag_event_project_type',
              order=df_targets['tag_event_project_type'].value_counts().index,
              palette="viridis")
plt.title("Распределение tag_event_project_type")
plt.xlabel("Tag Event Project Type")
plt.ylabel("Количество")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 6. Распределение по tag_event_project_region
# -----------------------------
plt.figure()
sns.countplot(data=df_targets, x='tag_event_project_region',
              order=df_targets['tag_event_project_region'].value_counts().index,
              palette="magma")
plt.title("Распределение tag_event_project_region")
plt.xlabel("Tag Event Project Region")
plt.ylabel("Количество")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 7. Анализ пользователей с несколькими событиями: распределение количества событий на пользователя
# -----------------------------
user_events = df_targets.groupby('smt_decision_makerid').size().reset_index(name='event_count')

plt.figure()
sns.histplot(user_events[user_events['event_count'] < 100], bins=30, kde=True, color='olive')
plt.title("Распределение количества событий на пользователя")
plt.xlabel("Количество событий")
plt.ylabel("Количество пользователей")
plt.show()
```

```python
df_targets['report_date'] = pd.to_datetime(df_targets['report_date'], errors='coerce')

user_events = (df_targets
               .sort_values(by=['smt_decision_makerid', 'report_date'])
               .groupby('smt_decision_makerid')['event']
               .agg(list)
               .reset_index(name='event_chain'))

user_events['event_count'] = user_events['event_chain'].apply(len)


user_events['chain_str'] = user_events['event_chain'].apply(lambda x: " -> ".join(x))
chain_counts = user_events['chain_str'].value_counts().reset_index()
chain_counts.columns = ['event_chain', 'count']

print("\nТоп-10 уникальных цепочек событий:")
print(chain_counts.head(10))

# проскорили для сейла получили вероятность низкую, выбрали действие
# звонок встреча бронь продажа
# кто не совершал действия
# оставили целевой интерес и большей действий не было
# id и репорт дата (сделать)
# действия на сайте -
# роудмап + описание подхода в текстовом виде
# id юзера и репорт дата. Четыре события и фичи для каждого клиента
```

```python
user_events_exploded = user_events[['smt_decision_makerid', 'event_chain']].explode('event_chain')
user_events_exploded['stage'] = user_events_exploded.groupby('smt_decision_makerid').cumcount() + 1

stage_event_counts = (user_events_exploded
                      .groupby(['stage', 'event_chain'])
                      .size()
                      .reset_index(name='count'))
print("\nРаспределение событий по этапам цепочки:")
print(stage_event_counts)

plt.figure(figsize=(12, 8))
sns.barplot(data=stage_event_counts[stage_event_counts['stage'] <= 5],
            x='stage', y='count', hue='event_chain', palette="Set2")
plt.title("Распределение типов событий по этапам цепочки (Stage 1-5)")
plt.xlabel("Этап цепочки")
plt.ylabel("Количество событий")
plt.legend(title="Тип события")
plt.show()
```

```python
df_targets['report_date'] = pd.to_datetime(df_targets['report_date'], errors='coerce')

df_targets = df_targets.sort_values(by=['smt_decision_makerid', 'report_date'])

user_events = (df_targets.groupby('smt_decision_makerid')['event']
               .agg(list)
               .reset_index(name='event_chain'))

sale_chains = user_events[user_events['event_chain'].apply(lambda events: 'sale' in events)].copy()

sale_chains['chain_str'] = sale_chains['event_chain'].apply(lambda events: " -> ".join(events))

chain_counts = sale_chains['chain_str'].value_counts().reset_index()
chain_counts.columns = ['chain', 'count']
print("Топ-10 уникальных цепочек, приводящих к sale:")
print(chain_counts.head(10))
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='chain', data=chain_counts.head(10), palette='viridis')
plt.title("Топ-10 цепочек, приводящих к sale")
plt.xlabel("Количество пользователей")
plt.ylabel("Цепочка событий")
plt.show()

def sale_position(events):
    try:
        return events.index('sale') + 1  # +1, чтобы нумерация начиналась с 1
    except ValueError:
        return None

sale_chains['sale_position'] = sale_chains['event_chain'].apply(sale_position)

plt.figure(figsize=(10, 6))
sns.countplot(x='sale_position', data=sale_chains, palette='magma')
plt.title("Распределение позиции 'sale' в цепочках")
plt.xlabel("Позиция 'sale' в цепочке")
plt.ylabel("Количество пользователей")
plt.show()
```
